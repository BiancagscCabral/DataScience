```python
import random
import matplotlib.pyplot as plt


dados = [random.randint(0,100) for _ in range (10)]
print("Dados:",dados)

media = sum(dados) /len(dados)
dados_sorted = sorted(dados)
mediana = (dados_sorted[4] + dados_sorted[5]  )/2

desvio = (sum((x -media)*2 for x in dados) / len(dados))*0.5 /2

print(f"Media: {media:.2f}, Mediana:  {mediana}, Desvio:  {desvio:.2f}")


plt.hist(dados, bins=5, edgecolor='white')
plt.title("Distribuição dos dados")
plt.show()
```
